Data cleaning

```python
# df.isna()  #.isna() only checks whether the particular value is missing or null or nan or not available
df.isna().sum()  #.sum() gives the total count of true values of .isna() and the output gives the total number null or missing 
                #values in the dataframe in a particular column
```

```python
# df.fillna(df.mean(),inplace=True) this will directly fill the missing values with mean values for all the columns but
#instead of this method columns are cleanaed individually
df['math score'].fillna(df['math score'].mean(),inplace=True)  

#inplace=True is used for applying any changes made directly to the dataframe , if it is now written then the changes appear 
#a copy of the dataframe
```

```python
df['reading score'].fillna(df['reading score'].mean(),inplace=True) 
```

```python
df['writing score'].fillna(df['writing score'].mean(),inplace=True) 
```

```python
df.isna().sum() #now all the columns are cleaned and free of missing values
```

Printing rows and columns with loc and iloc
the main difference between loc and iloc is that loc uses label-based indexing, whereas iloc uses integer-based indexing. loc allows you to access data using labels or boolean indexing, while iloc enables you to access data using integer positions. 


```python
df.loc[60:80,['math score']] #in .loc 60 and 80 both will be considered
```

```python
df.loc[60:80,['math score','gender']]
```

```python
df.iloc[60:80,[2,6,7]] #printing individual columns
```

```python
df.iloc[60:80,2:6] #printing range of columns
```

```python
df['math score'].skew()  #prints the skewness value
#skewness- It provides information about the shape of the distribution. 
#the skewness refers to the asymmetry of the distribution of the data points along the horizontal axis.
#positive,negative and zero skewness
# Skewness = (3 * (mean - median)) / standard deviation  -> pearson coefficient of skewness
```

```python
df['math score'].kurt() #kurtosis of the value
# Kurtosis-It provides information about the presence of extreme values or outliers in the data and the concentration
# of data around the mean.
#leptokurtic, mesokurtic, platykurtic
```

```python
df['math score'].var()  #variance
#variance-Variance is a statistical measure that quantifies the spread or dispersion of a set of data points. 
# It measures how far individual data points in a dataset are spread out from the mean or average value
```

Converting categorical data to quantitative data

```python
encoder = LabelEncoder() 
df['gender']= encoder.fit_transform(df['gender'])
df['lunch']=encoder.fit_transform(df['lunch'])
```
